# NYC high schools SAT results analysis

This project is conducted in order to analyze SAT results of high schools in New Your City and visualize geographical distribution of the best schools based on the SAT results.

The following questions will be answered during the analysis:
- **Students of which class size tend to perform better on SAT?** *This questions comes from a statement on small class benifits where researchers advocate a general idea that smaller classes are more effective. I will try to proof or bust this idea by exploring NYC schools datasets.*
- **Which borough (area) has the best schools based on SAT results?**

Additionaly, I will try to visualize the answer for the second question!

In this project, I will use a set of datasets from NYC data containing information about:
- AP test results (ap_2010.csv). AP test is used as means of assessing how much high school students learned in the class and whether they meet criteria for college credits (or enrollment).
- Class size (class_size_2010.csv). Contains information on average class sizes for each school
- Demographics (demographics_2010.csv). Annual school accounts of NYC schools population
- Graduation (graduation_2010.csv). Graduation outcome for each school
- Directory (directory_2014.csv). Contains information on high schools directories
- SAT results (sat_2012.csv). Average SAT results for each school
- Survey (survey_all.txt). Data on surveys from all schools
- Survey district 75 (survey_d75.txt). Data on surveys from district 75 schools

During the inspection, I found out that the ```class_size``` dataset does not have a ```DBN``` column. However, I also noticed that a DBN value is a combination of ```CSD``` and ```SCHOOL CODE``` values. I will add a ```DBN``` column to the ```class_size``` dataset by summing up those columns. In order to do that, I need to add zeros to the single digits and make it string.

````python
data["class_size"].head()
````

````python
data["class_size"]["padded_csd"] = data["class_size"]["CSD"].apply(lambda x: str(x).zfill(2) if len(str(x))<2 else str(x))
data["class_size"]["DBN"] = data["class_size"]["padded_csd"] + data["class_size"]["SCHOOL CODE"]

data["class_size"]["DBN"].head()
````

Before combining data together, I need to convert values of different SAT sections to numeric values and sum them up to get overall SAT results

````python
data["sat_results"]["SAT MATH AVG. SCORE"] = pd.to_numeric(data["sat_results"]["SAT MATH AVG. SCORE"], errors="coerce")
data["sat_results"]["SAT CRITICAL READING AVG. SCORE"] = pd.to_numeric(data["sat_results"]["SAT CRITICAL READING AVG. SCORE"], errors="coerce")
data["sat_results"]["SAT WRITING AVG. SCORE"] = pd.to_numeric(data["sat_results"]["SAT WRITING AVG. SCORE"], errors="coerce")
# Sum up all the respective values
data["sat_results"]["SAT_SCORE"] = data["sat_results"]["SAT MATH AVG. SCORE"] + data["sat_results"]["SAT CRITICAL READING AVG. SCORE"] + data["sat_results"]["SAT WRITING AVG. SCORE"]

data["sat_results"]["SAT_SCORE"].head()
````

Before I proceed to the next step, I will get longtitude and latitude values from the ```LOCATION 1``` column and assign it as 2 separate values

````python
import re

def find_lat(string):
    regex = "\(.+\)"
    data = re.findall(regex, string)
    result = data[0].split(",")[0].replace("(","")
    return result
````

````python
data["hs_directory"]["LAT"] = data["hs_directory"]["LOCATION 1"].apply(find_lat)

data["hs_directory"]["LAT"].head()
````

````python
def find_lon(string):
    regex = "\(.+\)"
    data = re.findall(regex, string)
    result = data[0].split(",")[1].replace(")","")
    return result
````

````python
data["hs_directory"]["LON"] = data["hs_directory"]["LOCATION 1"].apply(find_lon)

data["hs_directory"]["LON"].head()
````

**First of all, I will draw a map of NYC including rivers, coastlines and boundaries**

````python
combined["LAT"] = pd.to_numeric(combined["LAT"])
````

````python
combined["LON"] = pd.to_numeric(combined["LON"])
````

````python
plt.figure(figsize=(10,10))

m = Basemap (
    projection = 'merc',
    llcrnrlat = 40.496044,
    urcrnrlat = 40.915256,
    llcrnrlon = -74.255735,
    urcrnrlon = -73.700272,
    resolution = "i")

m.drawmapboundary(fill_color="#b1b3b1")
m.drawcoastlines(color="#6D5F47", linewidth=.4)
m.drawrivers(color="#6D5F47", linewidth=.4)

longitudes = combined["LON"].tolist()
latitudes = combined["LAT"].tolist()
m.scatter(longitudes, latitudes, s=20, zorder=2, latlon=True)
plt.show()
````

From the map above I can see that school density is highest in Manhattan (the top of the map) and lower in Brooklyn, the Bronx, Queens and State Islands. However, this map only shows geolocation of schools, it doesn't show any really meaningful information. I am gonna visualize the percentage of English learners by area.

````python
plt.figure(figsize=(10,10))

m_eng = Basemap (
    projection = 'merc',
    llcrnrlat = 40.496044,
    urcrnrlat = 40.915256,
    llcrnrlon = -74.255735,
    urcrnrlon = -73.700272,
    resolution = "i")

m_eng.drawmapboundary(fill_color="#b1b3b1")
m_eng.drawcoastlines(color="#6D5F47", linewidth=.4)
m_eng.drawrivers(color="#6D5F47", linewidth=.4)

longitudes = combined["LON"].tolist()
latitudes = combined["LAT"].tolist()
m_eng.scatter(longitudes, latitudes, s=20, zorder=2, latlon=True, c=combined["ELL_PERCENT"], cmap="summer")
plt.show()
````
